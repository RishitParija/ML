```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = pd.concat([delivery_df,loop_df])
```

```python
#LETS IDENTIFY THE NULL PROBLEMS
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
#AS WE SEE THE PATTERN OF NAMES WE CAN EXTRACT THE !ST WORD OF THE NAME AND PUT IT IN CITY IN PLACE OF NULL VALUES
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
df = df[df['city'].isin(eligible_cities)]
df
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
df
```

```python
    #BASICALLY HERE WE ARE CREATING THE COLUMN OF BALLS BOWLED
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
#CREATING COLUMN OF BALLS LEFT TO BE BOWLED
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
backup=df
backup
```

```python
#CREATING COLUMN FOR PLAYERS DISMISSED AND WICKETS LEFT
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:1 if x!='0' else 0)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
df
```

```python
#CURRENT RUN RATE COLUMN MAKING
df['crr'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
#IMPORTANT CELL TO CALCIULATE RUNS IN LAST 5 OVERS BY USING ROLLING WINDOW METHOD
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')#TO EXTRACT FINAL SCORE
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
final_df.dropna(inplace=True)
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])# TO SHUFFLE SO AS TO AVOID BIAS BEING DEVELOPED
final_df.sample(2)
#HERE WE FINALLY END WITHE PROCESS OF MAKING COLUMSS PERFECT FOR OUR NEEDS NOW WE WILL BEGIN BUILDING THE MODEL
```

```python
#HERE WE BEGIN CODING THE MODEL
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#AS WE HAVE 3 STRING COLUMNS WE HAVE TO ONE HOT ENCODE THEM

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
# CREATING A PIPELINE
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
#EXPORTING THE MODEL
pickle.dump(pipe,open('pipe.pkl','wb'))
```
